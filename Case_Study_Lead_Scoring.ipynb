###Declaring Functions

```python
# Get percentage of null values
def get_null_pct(dataframe1):
  return round((100/(dataframe1.shape[0]/dataframe1.isnull().sum())).sort_values(ascending=False), 2)
```

```python
#function to get outliers for a column
def get_outliers(dataframe1, column_name):

  if dataframe1[column_name].dtype == int or dataframe1[column_name].dtype == float:
    Q1 = dataframe1[column_name].quantile(0.25)
    Q3 = dataframe1[column_name].quantile(0.75)
    IQR = Q3 - Q1

    out_bottom = dataframe1[column_name] < (Q1-1.5*IQR)
    out_up = dataframe1[column_name] > (Q3+1.5*IQR)

    outliers = dataframe1[column_name][out_bottom | out_up]

    return outliers

  else:
    return "No outliers"
```

```python
def get_outliers_percentage(dataframe1, column_name):

  outliers = get_outliers(dataframe1, column_name)

  if type(outliers) == str:
    return outliers
  elif outliers.shape[0] == 0:
    return 0
  elif dataframe1.shape[0] > 0:
    return round(100/(dataframe1.shape[0]/outliers.shape[0]), 2)
```

```python
# Converting all the values to lower case
df = df.applymap(lambda s:s.lower() if type(s) == str else s)
```

As observed, there are multiple columns with only a single value. It is crucial to remove these columns for efficient analysis.

```python
# Removing columns with unique/single value

columns = []

for col, uniq_vals in df.nunique().items():
  if uniq_vals == 1:
    columns.append(col)

print("Dropping unique valued columns:\n")
print("\n".join(columns))

df1 = df.drop(columns, axis=1)
```

```python
# Checking the percentage of missing/null values
get_null_pct(df1)
```

We should remove columns with more than 40% missing values to ensure our data analysis is not skewed by unreliable or incomplete information.

```python
# Removing columns with 40% or more null values

columns = []

for col, null_pct in get_null_pct(df1).items():
  if null_pct >= 40:
    columns.append(col)

print("Dropping columns having 40% or more null values:\n")
print("\n".join(columns))

df2 = df1.drop(columns, axis=1)
```

```python
# Removing the columns that are not required for analysis

df2 = df2.drop(['Prospect ID', 'Lead Number'], axis=1)
```

```python
# Rechecking the percentage of missing values
get_null_pct(df2)
```

To retain crucial data in the columns with many null values, we'll replace NaN with 'N/A' instead of removing rows. This ensures data completeness for analysis, and if needed, we can handle or discard these values during model processing.

```python
# Replacing null values with dummy value for the columns having 20% or more null values

for col, null_pct in get_null_pct(df2).items():
  if null_pct >= 20:
    print("Column: {} | Replacing null values with dummy value...".format(col))
    df2[col] = df2[col].fillna('n/a')
```

##Plotting the ROC Curve

```python
def draw_roc(actual, probs):

    fpr, tpr, thresholds = metrics.roc_curve(actual, probs, drop_intermediate=False)
    auc_score = metrics.roc_auc_score(actual, probs)
    plt.figure(figsize=(5, 5))
    plt.plot(fpr, tpr, label='ROC Curve (area = %0.2f)' % auc_score)
    plt.plot([0, 1], [0, 1], 'g--')
    plt.xlim([-0.05, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend(loc="best")
    plt.show()
```

```python
draw_roc(y_train_pred_final['Converted'], y_train_pred_final['Convert_Prob'])
```

###Optimal Cutoff Point

```python
# Creating columns with different probability cutoffs
numbers = [float(x)/10 for x in range(10)]

for i in numbers:
    y_train_pred_final[i]= y_train_pred_final['Convert_Prob'].map(lambda x: 1 if x > i else 0)

y_train_pred_final.head()
```

```python
# Calculating accuracy, sensitivity, and specificity for different probability cutoffs.

cutoff_df = pd.DataFrame(columns = ['prob','accuracy','sensi','speci'])

for n in numbers:
    cmtrx = metrics.confusion_matrix(y_train_pred_final['Converted'], y_train_pred_final[n])

    total = sum(sum(cmtrx))

    accuracy = (cmtrx[0,0]+cmtrx[1,1])/total

    speci = cmtrx[0,0]/(cmtrx[0,0]+cmtrx[0,1])
    sensi = cmtrx[1,1]/(cmtrx[1,0]+cmtrx[1,1])

    cutoff_df.loc[n] =[n, accuracy, sensi, speci]

print(cutoff_df)
```

```python
# Let's plot accuracy, sensitivity, and specificity for different probabilitie cutoffs

cutoff_df.plot.line(x='prob', y=['accuracy','sensi','speci'])

plt.show()
```
